```python
# inladen bibliotheken en dataset
import scipy.stats as st
from scipy import stats
from scipy.stats import chi2_contingency 
import statsmodels.api as smapi 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
etmgeg = pd.DataFrame(pd.read_csv('etmgeg2.csv'))
```

Sinds 1949, na de troonsbestijging van Juliana, wordt Koninginnedag
gevierd op 30 april. Nadat Willem-Alexander koning was geworden werd
Koningsdag vanaf 2014 op 27 april gevierd. Hoewel dat dus maar drie
dagen eerder was, ontdekte de weervrouw dat het op Koningsdag echt
kouder is dan op Koninginnedag, meer dan 1 °C.  
***a) Klopt de bewering van de weervrouw statistisch gezien? Met andere woorden: was het op Koninginnedag significant warmer dan op Koningsdag?
Clichés hebben vaak een kern van waarheid. Zo wordt er wel beweerd dat een nieuw staatshoofd ook een andere wind laat waaien. Met onze data kunnen we vermoeden dat onderzoeken.***

Voor deze bewering gebruiken wij de ongepaarde T-toets, omdat de temperaturen op de verschillende dagen onafhankelijk van elkaar zijn.  
De hypothesen die wij hierbij hebben opgesteld zijn:
* Nulhypothese (H0): De gemiddelde temperatuur op Koninginnedag is gelijk aan de gemiddelde temperatuur op Koningsdag.  
* Alternatieve hypothese (Ha): De gemiddelde temperatuur op Koninginnedag is hoger dan de gemiddelde temperatuur op Koningsdag.  

```python
# we maken drie subdatasets
koningsdag_willem = etmgeg[(etmgeg['Staatshoofd'] == 'Willem-Alexander') & (etmgeg['Datum'].str.startswith('27-4'))]
koningsdag_juliana_beatrix = etmgeg[((etmgeg['Staatshoofd'] == 'Juliana') & (etmgeg['Datum'].str.startswith('30-4'))) |
                                    ((etmgeg['Staatshoofd'] == 'Beatrix') & (etmgeg['Datum'].str.startswith('30-4')))]
koningsdag = pd.concat([koningsdag_willem, koningsdag_juliana_beatrix])

#we voegen 2 koninginnen samen en zorgen voor duidelijkheid en gemak.
koningsdag['Staatshoofd'] = koningsdag['Staatshoofd'].replace(['Juliana', 'Beatrix'], 'koninginnendag')
koningsdag['Staatshoofd'] = koningsdag['Staatshoofd'].replace('Willem-Alexander', 'koningsdag')

#we selecteren de temperatuurgegevens
temp_koning = koningsdag_willem['TempGem']
temp_koningin = koningsdag_juliana_beatrix['TempGem']

#we plotten de boxplot
sns.boxplot(x='Staatshoofd', y='TempGem', data=koningsdag)
plt.xlabel('Staatshoofd')
plt.ylabel('Gemiddelde Temperatuur op Koningsdag')
plt.title('Gemiddelde Temperatuur op Koningsdag per Staatshoofd')
plt.show()

#we voeren de ongepaarde t-toets uit
stats.ttest_ind(temp_koning, temp_koningin, equal_var=False, alternative='less')
```

**Conclusie:** Op basis van de p-waarde die we hebben verkregen uit de ongepaarde t-toets, kunnen we de nul-hypothese verwerpen. Dit houdt in dat de gemiddelde temperatuur op Koninginnedag hoger is als de gemiddelde temperatuur op Koningsdag. De p-waarde die we hebben verkregen uit de t-toets is kleiner als het significantieniveau die we hebben gehanteerd.

***b) Waait de wind bij een ander staatshoofd ook echt uit een andere richting? Met andere woorden: zijn staatshoofd en windrichting onafhankelijke stochasten?***

Voor deze bewering gebruiken wij de chi-kwadraattoets, omdat we kijken naar twee categorische variabelen.  
De hypothesen die wij hierbij hebben opgesteld zijn:
* Nulhypothese (H0): Staatshoofd en windrichting zijn onafhankelijk van elkaar.  
* Alternatieve hypothese (Ha): Staatshoofd en windrichting zijn afhankelijk.

```python
# We maken een kruistabel
staat_wind = pd.crosstab(etmgeg['Staatshoofd'], etmgeg['WindrCat9']) 

# We voeren de chi-kwadraattoets uit
chi2_stat, p_value, dof, expected = chi2_contingency(staat_wind)
chi2_contingency(staat_wind) 
```

**Conclusie:** De p-waarde van de chi-kwadraattoets is groter dan 0.05, wat betekent dat we de nuhypothese niet kunnen verwerpen. Dit duidt erop dat de staatshoofd en windrichting afhankelijk zijn.

In april en augustus is er zelden vorst aan de grond. Er is wel elke dag een
minimumtemperatuur en een maximumtemperatuur. Uiteraard is de
maximumtemperatuur hoger dan de minimumtemperatuur. Maar wanneer
treden die op?  
***c) Het vermoeden is dat de minimumtemperatuur eerder op de dag
plaatsvindt dan de maximumtemperatuur. Klopt dit vermoeden?***

Om te bepalen of het vermoeden klopt dat de minimumtemperatuur eerder op de dag plaatsvindt dan de maximumtemperatuur, maken wij gebruik van een gepaarde t-toets. Deze toets is geschikt omdat je twee metingen per dag hebt die op dezelfde dag zijn genomen en dus aan elkaar gekoppeld zijn.  
De hypothesen die wij hierbij hebben opgesteld zijn:  
* Nulhypothese (H0): De tijdstippen van de minimumtemperatuur en de maximumtemperatuur zijn gemiddeld genomen gelijk. Dit impliceert dat er geen systematisch verschil is in de tijdstippen waarop de minimum- en maximumtemperaturen optreden.  
* Alternatieve hypothese (Ha): De minimumtemperatuur treedt gemiddeld eerder op de dag op dan de maximumtemperatuur.

```python
#we verwijderen rijen waar 'TempMinUur' of 'TempMaxUur' NaN zijn
etmgeg = etmgeg.dropna(subset=['TempMinUur', 'TempMaxUur'])

#We bereken het verschil tussen de Nederlandse cijfers en Engelse cijfers om een verband te zoeken. 
etmgeg['Verschil'] =etmgeg.TempMaxUur-etmgeg.TempMinUur

# We plotten een grafiek om te zien of de verschillen normaal verdeeld zijn.
sns.displot(etmgeg, x='Verschil', kind="kde")
smapi.qqplot(etmgeg.Verschil, fit=True, line='r')
plt.show()
```

Uit de uitvoer van de Q-Qplot en het verschil uit te drukken in een normaalverdeling is te zien dat de waarnemingen niet normaal verdeeld zijn. Daarom hebben we gekozen voor een binomiale tekentoets.


```python
# We hebben variabelen gemaakt om de 'tekens' te berekenen.
posi = len(etmgeg[ ( etmgeg['TempMinUur']<etmgeg['TempMaxUur'] ) ] ) 
toti = len(etmgeg) - len(etmgeg[(etmgeg['TempMinUur']==etmgeg['TempMaxUur'])])

# We voeren de tekentoets uit. Hierbij is de aanname dat er geen verschil is, dus een p= 1/2.  
st.binomtest(posi,toti,p=0.5,alternative='greater')  
```

**Conclusie:** De p-waarde van de toets is groter dan de significantieniveau 0.05. Dit houdt in dat we de nulhypothese verwerpen, dus er is significant bewijs dat de minimumtemperatuur  eerder op de dag plaatsvindt dan de maximumtemperatuur.
